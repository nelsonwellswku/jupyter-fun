```python
import pandas as pd
```

```python
numbers = pd.Series(range(1,101))
numbers_squared = numbers ** 2
numbers_cubed = numbers ** 3
numbers_fourth = numbers ** 4
numbers_fifth = numbers ** 5
```

```python
numbers_df = pd.DataFrame({'Numbers': numbers, 'Squared': numbers_squared, 'Cubed': numbers_cubed, 'Fourth': numbers_fourth, 'Fifth': numbers_fifth})
```
